**Вывод формулы для субградиента функционала в прямой задаче SVM без ограничений**

Прямая задача SVM без ограничений:
$$ \min\limits_{\pmb{w}} \frac{1}{2} ||\pmb{w}||^2 + C \sum\limits_{n=1}^N \max\{0, 1 - y_n(\pmb{w}^T\pmb{x}_n + w_0)\}$$

1. Нужно привести эту задачу к следующему виду:
$$\sum\limits_{n=1}^N f_n(w_0, \pmb{w}, \pmb{x}_n)$$
$$ \min\limits_{\pmb{w}} \sum\limits_{n=1}^N \left[\frac{1}{2N} ||\pmb{w}||^2 + C \max\{0, 1 - y_n(\pmb{w}^T\pmb{x}_n + w_0)\} \right]$$

2. Для метода субградиентного спуска нужно найти $\\grad f_n$, но $f_n$ диффиренцируема всюду, кроме точек, в которых $1 - y_n(\pmb{w}^T\pmb{x}_n + w_0) = 0$, в таких точках приходится использовать субградиент. Так как $\max\{0, \alpha\} \geq k \alpha$ при $0 \leq k \leq 1$, субградиент $f_n$ при $1 - y_n(\pmb{w}^T\pmb{x}_n + w_0) = 0$ можно вычислить как $$\\grad_{\pmb{w}} f_n = \frac{\pmb{w}}{N} - Ck y_n \pmb{x}_n$$ далее выберем $k = 1$. Получаем следующую формулу для субградиента: $$\\grad_{\pmb{w}} f_n = \mathbb{I}_{y_n(\pmb{w}^T\pmb{x}_n + w_0) \geq 0} \times (\frac{\pmb{w}}{N} - C y_n \pmb{x}_n)$$ где $\mathbb{I}$ - индикаторная функция. Аналогично: $$ \frac{\partial f_n}{\partial w_0} = -C y_n \mathbb{I}_{y_n(\pmb{w}^T\pmb{x}_n + w_0) \geq 0} $$
3. Субградиент целевой функции вычисляется как сумма субградиентов $f_n$

````python
import numpy as np
import lab3
from importlib import reload
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',**{'family':'verdana'})
rc('text', usetex=False)
rc('text.latex',unicode=True)
rc('text.latex',preamble=u'\\usepackage[utf8]{inputenc}')
rc('text.latex',preamble=u'\\usepackage[russian]{babel}')
rc('xtick', labelsize=12)
%matplotlib inline
import sklearn.svm
import sklearn.grid_search
````

**Генерация модельных данных**

````python
def gen_data(size=100, dim=2, separable=False):
    mean = np.random.poisson(3, dim) + 1
    cov = np.eye(dim, dim)
    if separable:
        X1 = np.random.multivariate_normal(-mean, cov, size)
        X2 = np.random.multivariate_normal(mean, cov, size)
    else:
        X1 = np.random.multivariate_normal(mean, cov, size)
        X2 = np.random.multivariate_normal(mean + 3, cov, size)
    X = np.vstack((X1, X2))
    y = np.array([-1]*size+[1]*size)
    return (X, y)
````

**3) Реализовать процедуру поиска оптимального параметра C и ширины RBF ядра $\gamma$ с помощью кросс-валидации. Исследовать зависимость ошибки на валидационной выборке от значений этих параметров. Рассмотреть случаи хорошо и плохо разделимый выборок** 

````python
#очень плохо разделимая выборка
def gen_data2(size=100, dim=2):
    mean = np.random.poisson(3, dim) + 1
    cov = np.eye(dim, dim)
    X1 = np.random.multivariate_normal(mean, cov, size)
    X2 = np.random.multivariate_normal(mean + 1.5, cov, size)
    X = np.vstack((X1, X2))
    y = np.array([-1]*size+[1]*size)
    return (X, y)
````

````python
X, y = gen_data2()
(X_train, X_test, y_train, y_test) = sklearn.cross_validation.train_test_split(X, y, train_size=0.7)
````

Функция поиска оптимальных значений $C$ и $\gamma$ 

````python
def get_optimal_params(X_train, y_train):
    lab3.visualize(X_train, y_train)
    C = np.logspace(-10, 10, 21)
    gamma = np.logspace(-10, 10, 21)
    svm = sklearn.svm.SVC()
    gs = sklearn.grid_search.GridSearchCV(svm, {'C': C, 'gamma': gamma}, n_jobs=3, cv=4, verbose=True)
    gs.fit(X_train, y_train)
    print(gs.best_score_)
    return gs.best_params_
print(get_optimal_params(X_train, y_train))
````

Построение матрицы зависимости точности на валидационной выборке от парметров $C$ и $\gamma$ 

````python
M = []
for C in np.logspace(-10, 10, 21):
    M.append([])
    for gamma in np.logspace(-10, 10, 21):
        svm = sklearn.svm.SVC(C=C, kernel='rbf', gamma=gamma, tol=1e-6, max_iter=500)
        svm.fit(X_train, y_train)
        predictions = svm.predict(X_test)
        M[-1].append(np.sum(predictions == y_test)*1.0/y_test.size)
M = np.array(M)
````

````python
print(M.max(), np.logspace(-10, 10, 21)[M.argmax()//21], np.logspace(-10, 10, 21)[M.argmax()%21])
````

Визуализация матрицы зависимости точности на валидационной выборке от парметров $C$ и $\gamma$ 

````python
plt.matshow(M, cmap=matplotlib.cm.plasma) 
fig = plt.gcf() 
fig.set_size_inches(16, 8)
f = lambda x: '{0:.1g}'.format(x) 
plt.xticks(np.arange(21), list(map(f, list(np.logspace(-10, 10, 21)))), rotation=50)
plt.yticks(np.arange(21), list(map(f, list(np.logspace(-10, 10, 21)))), fontsize=12) 
plt.colorbar()
plt.ylabel(u'Значения $C$', fontsize=16) 
plt.xlabel(u'Значения $\gamma$', fontsize=16) 
plt.title(u'Зависимость точности на валидационной выборке от параметров $C$ и $\gamma$\n\n', fontsize=16)
plt.show()
````

Но с другой стороны это очень близко к максимальному значению в матрице, достигаемому при параметрах $C = 10^4$, $\gamma = 0.001$

**5) Исследовать, как размер подвыборки, по которой считается субградиент, в методе стохастического субградиентного спуска влияет на скорость сходимости метода и на точность решения**

````python
reload(lab3)

f, ((x)) = plt.subplots(1, 3, sharex=False, sharey=False, figsize=(15, 5))
for i in range(3):
    X, y = gen_data(100, 2)
    res = lab3.svm_qp_primal_solver(X, y, 1)
    ans = lab3.compute_primal_objective(X, y, res['w0'], res['w'], 1)
    
    res1 = lab3.svm_subgradient_solver(X, y, 1, max_iter=1000, size = 10)
    obj_curve1 = res1['objective_curve']
    t1 = res1['time']
    res2 = lab3.svm_subgradient_solver(X, y, 1, max_iter=1000, size = 33)
    obj_curve2 = res2['objective_curve']
    t2 = res2['time']
    res3 = lab3.svm_subgradient_solver(X, y, 1, max_iter=1000, size = 50)
    obj_curve3 = res3['objective_curve']
    t3 = res3['time']
    res4 = lab3.svm_subgradient_solver(X, y, 1, max_iter=1000)
    obj_curve4 = res4['objective_curve']
    t4 = res4['time']
    print('\n Выборка №', i+1)
    print(' T(1/10 L) = {0:5.2f}\n T(1/3  L) = {1:5.2f}\n T(1/2  L) = {2:5.2f}\n T(1    L) = {3:5.2f}'.format(t1, t2, t3, t4))
    x[i].semilogy(np.arange(obj_curve1.size), np.abs(obj_curve1 - ans), label=u'$1/10 L$', linewidth=2)
    x[i].semilogy(np.arange(obj_curve2.size), np.abs(obj_curve2 - ans), label=u'$1/3 L$', linewidth=2)
    x[i].semilogy(np.arange(obj_curve3.size), np.abs(obj_curve3 - ans), label=u'$1/2 L$', linewidth=2)
    x[i].semilogy(np.arange(obj_curve4.size), np.abs(obj_curve4 - ans), label=u'$1 L$', linewidth=2)
    x[i].set_xlabel(u"Количество итераций \n Выборка №"+str(i+1), fontsize=14)
x[1].legend(loc=1, bbox_to_anchor=(1.2, 1.3), fontsize = 14, ncol=4)
x[0].set_ylabel(u"Отклонение от ответа", fontsize=14)
plt.show()
````

Для наглядности было взято число итераций = 1000, и небольшая выборка(по 100 объектов в классе, двумерное признаковое пространство).

**По полученным графикам можно сделать следующие выводы:**
* Скорость сходимости очень сильно зависит от выбора значений $\alpha$ и $\beta$ ($\alpha = 0.1$, $\beta = 0.7$)
* Неожиданно хорошо работает метод субградиентного спуска с долей подвыборки = $1/10$
* Дольше всего метод субградиентного спуска работает без стохастической оптимизации, но при хорошем выборе $\alpha$ и $\beta$ иногда он сходится за небольшое количество итераций, что его сильно ускоряет.
