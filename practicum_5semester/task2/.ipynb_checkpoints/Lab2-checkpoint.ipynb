#####4) Исследовать, каким точным алгоритмом поиска ближайших соседей следует пользоваться в различных ситуациях.

Функция **block_search** ищет ближайших соседей блоками, то есть делает запросы ближайших соседей для 
первых $N$ тестовых объектов, затем для следующих $N$, и так далее, и в конце объединяет полученные 
результаты.

При поиске $k$ ближайших соседей некоторые методы(brute, собственная реализация) строят в памяти матрицу попарных расстояний обучающей выборки и тестовой выборки, эта матрица может быть очень большой, поэтому нельзя выбирать параметр $N$ слишком большим. 

Но при этом ясно, что объединение полученных результатов достаточно времязатратная операция, значит нельзя выбирать параметр $N$ и слишком маленьким.

Опыты показали, что $N$ порядка $10^3$ уже работает быстрее меньших порядков и ещё не вызывает проблем с памятью. 

````python
def block_search(x_test, kneighbors, N = 1500):
    i = 0
    while i < x_test.shape[0]:
        x = x_test[i:min(i + N, x_test.shape[0])]
        if (i == 0):
            res = kneighbors(x)
        else:
            new_res = kneighbors(x)
            res = (np.vstack((res[0], new_res[0])), np.vstack((res[1], new_res[1])))
        i += N
    return res
````

**(a)** Собственная реализация алгоритма поиска ближайших соседей на основе кода подсчёта евклидова 
расстояния между двумя множествами точек

````python
def my_kneighbors(train, test, k = 5):
    train = train.astype('float64')
    test = test.astype('float64')
    dist = np.sum(test**2, axis=1)[:, np.newaxis] + np.sum(train**2, axis=1)[np.newaxis, :] - 2 * test.dot(train.T)
    dist = np.sqrt(dist)
    indexes = np.zeros((dist.shape[0], k)).astype('int64')
    distances = np.zeros((dist.shape[0], k)).astype('float64')
    for i in range(k):
        idx = indexes[:, i] = np.argmin(dist, axis = 1)
        distances[:, i] = dist[np.arange(dist.shape[0]), idx]
        dist[np.arange(dist.shape[0]), idx] = np.inf
    return (distances, indexes)
````

#####Запуск эксперимента:
Требуется искать $5$ ближайших соседей в обучающей выборке для тестовой выборки (при этом ответы на тестовой выборке не используются). Метрика евклидова. Число признаков: $10$, $20$, $100$. Подмножество признаков выбирается один раз, случайно. Алгоритмы поиска ближайших соседей:

   **(a)** Собственная реализация на основе кода подсчёта евклидова расстояния между двумя множествами точек

   **(b)** sklearn.neighbors.NearestNeighbors(algorithm=’brute’)
   
   **(c)** sklearn.neighbors.NearestNeighbors(algorithm=’kd_tree’)
    
   **(d)** sklearn.neighbors.NearestNeighbors(algorithm=’ball_tree’)


````python
for cnt_features in [10, 20, 100]:
    features = np.arange(0, 28*28)
    np.random.shuffle(features)
    features = features[0:cnt_features]
    x_train = X_train[:, features]
    x_test = X_test[:, features]

    print('cnt_features = ', cnt_features)

    print('    my_knn')
    %time block_search(x_test, lambda a: my_kneighbors(x_train, a))

    print('    brute')
    KNN_v1 = sklearn.neighbors.NearestNeighbors(n_neighbors=5, algorithm="brute")
    %time KNN_v1.fit(x_train)
    %time block_search(x_test, KNN_v1.kneighbors)

    print('    kd_tree')
    KNN_v2 = sklearn.neighbors.NearestNeighbors(n_neighbors=5, algorithm="kd_tree")
    %time KNN_v2.fit(x_train)
    %time KNN_v2.kneighbors(x_test)

    print('    ball_tree')
    KNN_v3 = sklearn.neighbors.NearestNeighbors(n_neighbors=5, algorithm="ball_tree")
    %time KNN_v3.fit(x_train)
    %time KNN_v3.kneighbors(x_test)
    print()
    
````

 | $20$ признаков       | 59 s.         | 44 s.        | **21 s.**        | 2 min.         |
 | $100$ признаков      | 1 min. 6 s.   |  **56 s.**   | 6 min. 17 s.     | 6 min. 15 s.   |
 
 При $10$ признаках видно, что tree-алгоритмы работают быстрее, но при росте признаков Ball-tree работает заметно медленнее уже на $20$ признаках, а KD-tree на $100$. Из этого можно сделать вывод, что при небольшим количестве признаков следует использовать KD-tree, а при большом Brute, так как с ростом признаков время его работы менялось незначительно.
 
 **Это соответствует теоретическим соображениям:** 
 
 Так при небольшой размерности пространства($10-20$) kd-tree действительно находит ближайшего соседа за время порядка $O(\log{l})$, где $l$ - размер обучающей выборки. Экспериментально было установлено, что в пространствах большей размерности сложность поиска ближайшего соседа в kd-tree сильно ухудшается и приобретает линейный порядок сложности(Weber, Schek, Blott "A Quantitative Analysis and Performance Study for Similarity-Search Methods in High-Dimensional Spaces"). 
 
 Алгоритм brute также находит ближайшего соседа за линейное время $O(l)$, но благодаря оптимизациям в brute и дополнительным накладным расходам в kd-tree, на большой размерности пространства brute работает сильно лучше. 
 
 Ball-tree не зависит от размерности пространства, но обучение из-за этого усложняется. Так же при обучении нужно специально предобрабатывать данные, чтобы Ball-tree построил оптимальную обученную структуру. В данном случае предобработки данных не было и алгоритм показал себя не лучшим образом. 
 
 Собственная реализация, так же как и brute, ищет ближайшего соседа за линейное время, но она менее оптимизирована, следовательно на всех размерностях работает немного хуже brute.

#####5) Реализация генерации индексов обучающей и валидационной выборки для кросс-валидации с N фолдами.
Функция **my_cross_validation** по длине выборки $L$, количеству фолдов $N$ и номеру фолда $i$ возвращает набор индексов элементов, входящих в $i$-ый фолд и набор индексов всех остальных элементов.

````python
def my_cross_validation(L, N, i):
    permutation = np.arange(L)
    np.random.seed(N)
    np.random.shuffle(permutation)
    test = np.arange(i, L, N)
    train = np.delete(np.arange(L), test)
    return(permutation[train], permutation[test])
````

#####6) Реализация оценки точности метода k ближайших соседей по валидационной выборке для нескольких параметров ${\bf k}: [k_1, \dots , k_n],~k_1 < k_2 < \dots < k_n$.

````python
def precision(x_train, x_test, y_train, y_test, k, metr, weights=False, cnt_classes=10):
    KNN = sklearn.neighbors.NearestNeighbors(n_neighbors=k[-1], metric=metr, algorithm="brute")
    KNN.fit(x_train)
    (distances, indexes) = block_search(x_test, KNN.kneighbors)
    id_sort = np.argsort(distances, axis = 1)
    rows = np.reshape(np.arange(x_test.shape[0]), (x_test.shape[0], 1))
    distances = distances[rows, id_sort]
    indexes = indexes[rows, id_sort]
    y = y_train[indexes].astype('int32')
    w = np.ones(y.shape)
    if weights:
        w = 1 / (distances + 1e-4)
    rows = np.arange(y.shape[0])
    count = np.zeros((y.shape[0], cnt_classes))
    ans = np.zeros(k[-1])
    for i in range(k[-1]):
        count[rows, y[:, i]] += w[:, i]
        ans[i] = np.sum(y_test == np.argmax(count, axis = 1)) / y_test.size
    return ans[k - 1]
````

````python
import time
cnt_neighbors = 10
for metric in ['euclidean', 'cosine']:
    print()
    print("metric =", metric)
    for weight in [False, True]:
        print("weight = ", weight)
        T = 0
        P = np.zeros(cnt_neighbors)
        for i in range(3):
            print('i = ', i)
            (ix_train, ix_test) = my_cross_validation(mnist['data'].shape[0], 3, i)
            ix_test = ix_test
            target = mnist['target']
            data = mnist['data']
            t = time.clock()
            P += precision(data[ix_train], data[ix_test], 
                          target[ix_train], target[ix_test], 
                          np.arange(1, cnt_neighbors+1), metric, weight)
            T += time.clock() - t
            print("{:.3f} s".format(T))
        print()
        print(P / 3)
        print("{:.3f} s".format(T / 3))
        print()
````

#####9) Применить лучший алгоритм к исходной обучающей и тестовой выборке. 

````python
target = mnist['target']
data = mnist['data']
X_train = data[:60000]
X_test = data[60000:]
Y_train = target[:60000]
Y_test = target[60000:]
````

````python
K = 4
KNN = sklearn.neighbors.NearestNeighbors(n_neighbors=K, metric='cosine', algorithm="brute")
KNN.fit(X_train)
(distances, indexes) = block_search(X_test, KNN.kneighbors)
id_sort = np.argsort(distances, axis = 1)
rows = np.reshape(np.arange(X_test.shape[0]), (X_test.shape[0], 1))
distances = distances[rows, id_sort]
indexes = indexes[rows, id_sort]
y = Y_train[indexes].astype('int32')
w = 1 / (distances + 1e-4)
rows = np.arange(y.shape[0])
count = np.zeros((y.shape[0], 10))
for i in range(K):
    count[rows, y[:, i]] += w[:, i]
z = np.argmax(count, axis = 1)
P = np.sum(Y_test == z) / Y_test.size
````

````python
print("Precision = {:.2f}%".format(P*100))
````

````python
import re
docs = list(newsgroups.data)
N = len(docs)
all_words = []
for i in range(N):
    docs[i] = re.sub("\W+", ' ', docs[i].lower()).split(' ')
    all_words += docs[i]
all_words = list(set(all_words))
dict_words = dict(zip(all_words, range(len(all_words))))

indptr = [0]
indexes = []
data = []
for d in docs:
    for term in d:
        index = dict_words[term]
        indexes.append(index)
        data.append(1)
    indptr.append(len(indices))
csr = scipy.sparse.csr_matrix((data, indexes, indptr), shape=(N, len(all_words)), dtype=int)
````

```python 
docs[i] = re.sub("\W+", ' ', docs[i].lower()).split(' ') 
```

При приобразовании tfidf происходит нормировка длин векторов элементов, после которой все элементы будут иметь одинаковую длину. Если же  не вводить дополнительное пустое слово, то вектора, соответствующие пустым текстам, будут иметь нулевую длину, и ошибочно оказываться ближайшими к непустым.

#####17) Загрузить тестовую выборку (параметр subset метода fetch_20newsgroups). Применить лучший алгоритм к тестовой выборке. Сравнить точность с полученной по кросс-валидации.

````python
test = sklearn.datasets.fetch_20newsgroups(remove=('headers', 'footers', 'quotes'), subset='test')
````

````python
import re
test_docs = list(test.data)
M = len(test_docs)
for i in range(M):
    test_docs[i] = re.sub("\W+", ' ', test_docs[i].lower()).split(' ')

indptr = [0]
indices = []
data = []
for d in test_docs:
    for term in d:
        if term in dict_words:
            index = dict_words[term]
            indices.append(index)
            data.append(1)
    indptr.append(len(indices))
test_csr = scipy.sparse.csr_matrix((data, indices, indptr), shape=(M, len(all_words)), dtype=int)
test_t = sklearn.feature_extraction.text.TfidfTransformer()
test_tfidf = test_t.fit_transform(test_csr)
````

````python
K = 6
KNN = sklearn.neighbors.NearestNeighbors(n_neighbors=K, metric='cosine', algorithm="brute")
KNN.fit(tfidf)
(distances, indexes) = block_search(test_tfidf, KNN.kneighbors)
id_sort = np.argsort(distances, axis = 1)
rows = np.reshape(np.arange(test_tfidf.shape[0]), (test_tfidf.shape[0], 1))
distances = distances[rows, id_sort]
indexes = indexes[rows, id_sort]
y = Y_train[indexes].astype('int32')
w = 1 / (distances + 1e-4)
rows = np.arange(y.shape[0])
count = np.zeros((y.shape[0], 20))
for i in range(K):
    count[rows, y[:, i]] += w[:, i]
z = np.argmax(count, axis = 1)
P = np.sum(test.target == z) / test.target.size
````
